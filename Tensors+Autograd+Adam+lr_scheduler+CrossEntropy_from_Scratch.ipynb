```python
import math
class LinearClassification(torch.nn.Module):
  def __init__(self,input_dim, output_dim):

    super().__init__()
    std = math.sqrt(1/(input_dim+output_dim))

    self.weight = torch.nn.Parameter(torch.nn.init.trunc_normal_(torch.empty(output_dim,input_dim),std= std,a = -3*std,b = 3*std),requires_grad = True)
    self.bias = torch.nn.Parameter(torch.nn.init.trunc_normal_(torch.empty(output_dim), std=std, a= -1*std,b = std),requires_grad=True)

  def forward(self,x):
    return x@self.weight.T + self.bias

```

```python
def softmax(x,dim = -1):
  x_max = torch.max(x,dim= dim ,keepdim= True)[0]
  rescaled = x - x_max
  exp_rescaled = torch.exp(rescaled)
  exp_sum = torch.sum(exp_rescaled,dim = dim,keepdim =True)
  return exp_rescaled/exp_sum

def log_softmax(x,dim = -1):
  x_max = torch.max(x,dim= dim ,keepdim= True)[0]
  rescaled = x - x_max
  exp_rescaled = torch.exp(rescaled)
  exp_sum = torch.sum(exp_rescaled,dim = dim,keepdim =True)
  return rescaled - torch.log(exp_sum)


def cross_entropy(inputs,targets):
  negative_logits = -log_softmax(inputs,dim = -1)
  return torch.mean(torch.gather(negative_logits,-1, targets.unsqueeze(-1)))

def clip_gradient(parameters,max_norm):
  grads= [p.grad for p in parameters if p.grad is not None]
  if not grads:
      return

  norm = torch.tensor(0.0 , device = grads[0].device)

  for g in grads:

    norm+= torch.sum(g**2)

  norm = torch.sqrt(norm)


  clip_coeff= min(1.0, max_norm/(norm.item()+1e-6))

  for g in grads:
    g.mul_(clip_coeff)
```

```python
def get_cosine_lr(it, warmup_iters, cosine_iters,min_learning_rate, max_learning_rate):

  if it < warmup_iters:
    return max_learning_rate* (it/warmup_iters)
  if it > cosine_iters:
    return min_learning_rate

  decay =(it-warmup_iters)/(cosine_iters-warmup_iters)
  decay_rate = 0.5*(1+math.cos(math.pi*decay))
  return min_learning_rate+decay_rate*(max_learning_rate-min_learning_rate)
```

```python
from __future__ import annotations
from typing import Callable

class AdamW(torch.optim.Optimizer):
  def __init__(self, params, lr = 1e-3, betas = (0.9,0.999), eps = 1e-8, weight_decay = 0.0):
    defaults = dict(lr = lr, betas = betas, eps = eps, weight_decay = weight_decay)
    super().__init__(params,defaults)

  def step(self, closure: Callable | None =None):

    loss = None
    if not closure is None:
      loss = closure()

    for group in self.param_groups:
      for p in group['params']:
        if p.grad is None:
          continue
        grad = p.grad.data

        if grad.is_sparse:
          raise RuntimeError('AdamW does not support sparse gradients')

        # Correct way to access state for each parameter
        state = self.state[p]

        alpha = group['lr']
        beta1, beta2 = group['betas']
        eps = group['eps']
        weight_decay = group['weight_decay']
        t = state.get("t",1)

        alpha_t = alpha *math.sqrt((1-beta2**t)/(1-beta1**t))
        p.data -= alpha*weight_decay*p.data

        prev_mt = state.get('m',torch.zeros_like(grad))
        prev_vt = state.get('v',torch.zeros_like(grad))
        # print(prev_mt.shape,prev_vt.shape,grad.shape)
        mt = beta1*prev_mt + (1-beta1)* grad
        vt = beta2*prev_vt + (1-beta2)*(torch.square(grad))

        p.data -= alpha_t*(mt/(torch.sqrt(vt)+eps))

        state['m'] = mt
        state['v'] = vt
        state['t'] = t+1

    return loss
```

```python
def cross_entropy_backwards(inputs, targets):
  batch_size = inputs.shape[0]
  probs = softmax(inputs).to(torch.accelerator.current_accelerator())
  probs[torch.arange(batch_size),targets] -= 1
  return probs/batch_size

def backprop(dout,X, W):
  # Changed x to X to use the local GPU tensor
  dw = dout.T @ X
  db = torch.sum(dout,dim = 0)
  dx = dout @ W

  return dw,db,dx
```

```python
model = LinearClassification(X_train.shape[1],3).to(torch.accelerator.current_accelerator())
criterion = cross_entropy
optimizer = AdamW(model.parameters(),lr = 0.001)
```

```python
X_tensor = torch.tensor(X_train, dtype=torch.float32).to(torch.accelerator.current_accelerator())
y_tensor = torch.tensor(y_train,dtype = torch.long).to(torch.accelerator.current_accelerator())
```

```python
warmup_iter= 200
cosine_iter = 1500
min_learning_rate = 1e-6
max_learning_rate = 0.1

for it in range(2000):
  outputs = model(X_tensor)
  lr = get_cosine_lr(it,warmup_iter,cosine_iter,min_learning_rate,max_learning_rate)
  for g in optimizer.param_groups:
    g['lr'] = lr
  optimizer.zero_grad()
  logits = cross_entropy_backwards(outputs, y_tensor)

  dw, db, _ = backprop(logits, X_tensor, model.weight)
  model.weight.grad = dw
  model.bias.grad = db
  clip_gradient(model.parameters(), 1)
  optimizer.step()
```

```python
outputs = model(X_tensor)

y_pred = softmax(outputs)
y_pred = torch.argmax(y_pred,dim = -1)
accuracy = (y_pred == y_tensor).sum().item()/y_tensor.shape[0]
accuracy
```
